```python
from sklearn_pandas import DataFrameMapper as DFM
import pandas as pd
from sklearn.preprocessing import LabelBinarizer, StandardScaler, MinMaxScaler
```

```python
train = pd.DataFrame({'pet':      ['cat', 'dog', 'dog', 'fish', 'cat', 'dog', 'cat', 'fish'],
                      'sex':      ['M', 'F', 'F', 'F', 'M', 'M', 'F', 'M'],
                      'children': [4., 6, 3, 3, 2, 3, 5, 4],
                      'salary':   [90., 24, 44, 27, 32, 59, 36, 27]})

test = pd.DataFrame({'pet':      ['dog', 'cat', 'fish'],
                     'sex':      ['M', 'F', 'F'],
                     'children': [1.   ,  2   , 1.6],
                     'salary':   [10.0 ,  5   , 100]})
```

---

```python
mapper = DFM([('pet', LabelBinarizer())], df_out=True)
mapper.fit_transform(train)
```

---

```python
mapper = DFM([('pet', LabelBinarizer())], df_out=True, default=None)
mapper.fit_transform(train)
```

---

```python
mapper = DFM([(['children','salary'], MinMaxScaler()),
              ('sex'                , LabelBinarizer()),
              ('pet'                , LabelBinarizer())], 
             default=None,
             df_out=True)
mapper.fit_transform(train)
```

```python
('sex'                , LabelBinarizer()), \
('pet'                , LabelBinarizer())
```

```python
[(s, LabelBinarizer()) for s in ['sex', 'pet']]
```

```python
mapper = DFM([(['children','salary'], MinMaxScaler()),
              *[(s, LabelBinarizer()) for s in ['sex', 'pet']]],
             default=None,
             df_out=True)
mapper.fit_transform(train)
```
